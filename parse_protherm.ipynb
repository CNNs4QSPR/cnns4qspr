```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
file1 = open('protherm_secondary.txt','r')
```

```python
soup1 = bs(file1, 'html.parser')
```

```python
# this finds all the column headers
header_list = []
header_tags = soup1.find_all('tr')[0].find_all('th')
for tag in header_tags:
    header_list.append(tag.contents[0])
```

```python
protein_dict = {header:[] for header in header_list}
```

```python
for protein in soup1.find_all('tr')[1:]:
    data_tags = protein.find_all('td')
    for index, tag in enumerate(data_tags):
        try:
            protein_dict[header_list[index]].append(tag.contents[0])
        except:
            protein_dict[header_list[index]].append('empty')
```

```python
len(protein_dict['dG'])
```

```python
df = pd.DataFrame(protein_dict)
```

```python
# finall 
bool1 = df['PDB_wild'] != 'empty'
bool2 = df['Tm'] != 'empty'
bool3 = bool1 & bool2
```
